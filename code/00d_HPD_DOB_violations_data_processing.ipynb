```python
#convert to datetime
df2['InspectionDate']= pd.to_datetime(df2['InspectionDate'], errors='coerce')
```

```python
#remove incorrect dates
years = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]

df2_2012_2022 = df2[df2.year.isin(years)]

df2_2012_2022
```

```python
#all hpd violations - date filtered

df2_2012_2022.to_csv('hpd_2012_2022.csv')
```

```python

#count of inspections per census tract
hpd_per_census_tract = df2_2012_2022[['Borough','BoroID','CensusTract', 'InspectionDate']].groupby(['Borough','BoroID','CensusTract']).agg(['count'])


```

```python
hpd_per_census_tract
```

```python
hpd_per_census_tract.to_csv('hpd_per_census_tract3.csv')
```

```python

#count of inspections per BBL
hpd_per_BBL = df2_2012_2022[['BBL','InspectionDate']].groupby(['BBL']).agg(['count'])


```
